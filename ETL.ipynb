Se crea dataframe para cada hoja a analizar en el EDA, se realiza un breve ETL y se guardan

```python
hojas= ['Velocidad % por prov', 'Accesos Por Tecnología', 'Penetración-poblacion', 'Penetracion-hogares','Penetracion-totales', 'Totales Accesos por rango', 'Accesos por rangos','Ingresos' ]
velocidad_provincia = internet.parse('Velocidad % por prov')
totales_accesos_tecnologia = internet.parse('Totales Accesos Por Tecnología')
accesos_tecnologia_provincia=  internet.parse('Accesos Por Tecnología')
p_poblacion =  internet.parse('Penetración-poblacion')
p_hogares =  internet.parse('Penetracion-hogares')
p_totales =  internet.parse('Penetracion-totales')
total_accesos_velocidad = internet.parse('Totales Accesos por rango')
accesos_velocidad_provincia = internet.parse('Accesos por rangos')
ingresos=  internet.parse('Ingresos ')
```

```python
totales_accesos_tecnologia.drop('Periodo', axis=1, inplace=True)
p_totales.drop('Periodo', axis=1, inplace=True)
ingresos.drop('Periodo', axis=1, inplace=True)
```

```python
totales_accesos_tecnologia.head(1)
# p_totales.head(1)
# ingresos.head()
```

```python
# Pesos argentinos a dolares y se renombra columna
ingresos['Ingresos (miles de pesos)']= round(ingresos['Ingresos (miles de pesos)']*0.001002, 2)
ingresos.rename({'Ingresos (miles de pesos)': 'Ingresos (dolares)'}, axis=1, inplace=True)
#Cambio de error de tipeo 2033
ingresos['Año']=ingresos['Año'].replace(2033,2023)
```

```python
# elimina los * 
accesos_tecnologia_provincia.head()
```

```python
type(accesos_tecnologia_provincia['Año'][0])
```

```python
# Como columna Año y Trimestre e int, se transforma a str para extraer la fecha, se elimina espacios y despues se vuelve a convertir a int
accesos_tecnologia_provincia['Año'] = accesos_tecnologia_provincia['Año'].astype(str).str.replace(r'[^\d]', '', regex=True).astype('Int64')
accesos_tecnologia_provincia['Trimestre'] = accesos_tecnologia_provincia['Trimestre'].astype(str).str.replace(r'[^\d]', '', regex=True)
```

```python
accesos_tecnologia_provincia[accesos_tecnologia_provincia['Año']=='2019']
```

```python
# Union de p_poblacion, o_hogar, p_totales
# Merge entre p_hogares y p_poblacion basado en Año, Trimestre y Provincia
df_merged = pd.merge(p_hogares, p_poblacion, on=['Año', 'Trimestre', 'Provincia'], how='outer')

penetracion_df = pd.merge(df_merged, p_totales, on=['Año', 'Trimestre'], how='outer')

# Renombrar columnas para mayor claridad
penetracion_df = penetracion_df.rename(columns={
    'Accesos por cada 100 hogares_x': 'Accesos por cada 100 hogares (Provincia)',
    'Accesos por cada 100 hab_x': 'Accesos por cada 100 hab (Provincia)',
    'Accesos por cada 100 hogares_y': 'Accesos por cada 100 hogares (Total)',
    'Accesos por cada 100 hab_y': 'Accesos por cada 100 hab (Total)'
})

# penetracion_df.head()

# df_merged.head()
penetracion_df.head()
```
